```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


if (website == "indeed"):
    pd.options.display.html.use_mathjax = False
    
    jobTitle = jobTitle.replace(" ","+")
    url = "https://www.indeed.com/jobs?q="+ jobTitle + "\&" + "l=" + location
    
    
    #Identify Yourself
    headers = {"user-agent": "IndeedBot/1.0"}
    page = requests.get(url, headers = headers)


    if page.status_code != requests.codes.ok :
        print("Request was not successful, status code:", page.status_code)
        exit()

    #Parse using Beautiful Soup
    soup = BeautifulSoup(page.content, 'html.parser')

    #Create a list of all jobs
    jobs = soup.find_all('div',{'class':'jobsearch-SerpJobCard'})

    #Declare lists
    title = []
    company = []
    location = []
    links = []
    

    #Go through all the jobs and save the necessary information in the lists
    for j in jobs:
        t = j.find('h2',{'class':'title'})
        t = t.get_text().strip()
        title.append(t)
        
        jobLink = j.find('a')
        link = "https://www.indeed.com" + jobLink.get('href')     
        links.append(link)

        c = j.find('span',{'class':'company'})
        c = c.get_text().strip()
        company.append(c)

        #location is either stored in a span tag or div tag therefore go through both
        l = j.find('span',{'class':'location'})
        try:
            l = l.get_text().strip()
        except AttributeError:
            l = j.find('div',{'class':'location'})
            l = l.get_text().strip()
        location.append(l)
  
```

```python
if (website == "linkedin"):
    pd.options.display.html.use_mathjax = False
    
    jobTitle = jobTitle.replace(" ","-")
    
    url = "https://www.linkedin.com/jobs/"+ jobTitle + "-jobs-" + location
    
    #Identify Yourself
    headers = {"user-agent": "LinkedinBot/1.0"}
    page = requests.get(url, headers = headers)


    if page.status_code != requests.codes.ok :
        print("Request was not successful, status code:", page.status_code)
        exit()

    #Parse using Beautiful Soup
    soup = BeautifulSoup(page.content, 'html.parser')
    
    #Create lists
    title = []
    company = []
    location = []
    links = []
    
    #Both the title of the job and the link to the job page are in the same <a> tag.
    jobTitle = soup.find_all('a',{'class':'result-card__full-card-link'})
    for job in jobTitle:
        link = job.get('href')   
        t = job.get_text().strip()
        title.append(t)
        links.append(link)
        
    employerList = soup.find_all('a',{'class':'result-card__subtitle-link job-result-card__subtitle-link'})
    for employer in employerList:
        e = employer.get_text().strip()
        company.append(e)
        
    locationList = soup.find_all('span',{'class':'job-result-card__location'})
    for l in locationList:
        l = l.get_text().strip()
        location.append(l)

```

```python
if (website == "monster"):
    pd.options.display.html.use_mathjax = False
    
    jobTitle = jobTitle.replace(" ","-")
    
    url = "https://www.monster.com/jobs/search/?q="+ jobTitle + "&where=" + location
          
    
    #Identify Yourself
    headers = {"user-agent": "monsterBot/1.0"}
    page = requests.get(url, headers = headers)


    if page.status_code != requests.codes.ok :
        print("Request was not successful, status code:", page.status_code)
        exit()

    #Parse using Beautiful Soup
    soup = BeautifulSoup(page.content, 'html.parser')
    
    #Create lists
    title = []
    company = []
    location = []
    links = []
    
    #Create a list of all jobs
    jobs = soup.find_all('div',{'class':'summary'})
    
    for j in jobs:
        t = j.find('h2',{'class':'title'})
        t = t.get_text().strip()
        title.append(t)
    
        c = j.find('div',{'class':'company'})
        c = c.get_text().strip()
        company.append(c)
        
        l = j.find('div',{'class':'location'})
        l = l.get_text().strip()
        location.append(l)
        
        link = j.find('a')
        link = link.get('href')   
        links.append(link)


    
```

```python
df= pd.DataFrame(data = {"Job Title":title, "Company Name":company, "Location":location, "Link":links})
df


```
